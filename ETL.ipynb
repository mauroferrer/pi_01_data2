```python
df2.replace('', np.nan, inplace=True)
```

```python
Iso = []
Name = []
for i in range(19):
    iso = 'iso_639_1' + str(i)
    nam = 'name' + str(i)
    Iso.append(iso)
    Name.append(nam)
    df5[iso] = pd.json_normalize(df5[i])['iso_639_1']
    df5[nam] = pd. json_normalize(df5[i])['name']
```

```python
data_concat = pd.concat([movie,df1,df2,df3,df4,df5], axis=1)
```

```python
data_concat = data_concat.drop(columns=['belongs_to_collection','spoken_languages', 'production_countries', 'genres', 'production_companies'])
```

```python
#### RELLENAMOS LOS VALORES NULOS CON 0 EN LAS COLUMNAS revenue Y budget
data_concat['revenue'] = data_concat['revenue'].replace(np.nan, 0)
data_concat['budget'] = data_concat['budget'].replace(np.nan, 0)
```

```python
#### ELIMINAMOS LOS VALORES NULOS DE LA COLUMNA release_date
data_concat['release_date'] = data_concat['release_date'].dropna()
```

```python
#### CAMBIAMOS A TIPO FECHA LA COLUMNA release_date Y CREAMOS LA COLUMNA release_year 
data_concat['release_date'] = pd.to_datetime(data_concat['release_date'], format= '%Y-%m-%d', errors='coerce')
data_concat['release_year'] = data_concat['release_date'].dt.year
```

```python
#### CREAMOS LA COLUMNA return CON LAS COLUMNA revenue Y budget
data_concat['budget'] = pd.to_numeric(data_concat['budget'], downcast = 'float', errors = 'coerce')
data_concat['return'] = (data_concat['revenue']).divide(data_concat['budget']).fillna(0)
data_concat['return'].replace([np.inf, -np.inf], np.nan, inplace=True)
data_concat['return'].fillna(0, inplace=True)
```

```python
#### ELIMINAMOS LAS COLUMNAS QUE NO SON RELEVANTES
data_concat = data_concat.drop(columns=['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'])
```

## TRANSFORMAMOS EL SEGUNDO DATASETS credits

```python
credit.head()
```

```python
#### DESANIDAMOS LA COLUMNA cast
credit['cast'] = credit['cast'].fillna('{}')
Df1 = pd.json_normalize(credit['cast'].map(eval))
Df1
```

```python
Df1[0][1]
```

```python
Character = []
Name = []
Gender = []
for i in range(313):
    char = 'character' + str(i)
    nam = 'name' + str(i)
    gen = 'gender' + str(i)
    Character.append(char)
    Name.append(nam)
    Gender.append(gen)
    Df1[char] = pd.json_normalize(Df1[i])['character']
    Df1[nam]  = pd.json_normalize(Df1[i])['name']
    Df1[gen]  = pd.json_normalize(Df1[i])['gender']
```

```python
col = list(range(313))
Df1.drop(columns=col, inplace=True)
```

```python
Df1['character']  =   Df1[Character].apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1)
Df1['name_actor'] =   Df1[Name]     .apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1)
Df1['genres_actor'] = Df1[Gender]   .apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1) 
```

```python
Df1.drop(columns=Character, inplace=True)
Df1.drop(columns=Name, inplace=True)
Df1.drop(columns=Gender, inplace=True)
Df1
```

```python
#### DESANIDAMOS LA COLUMNA crew
credit['crew'][0]
credit['crew'] = credit['crew'].fillna('{}')
Df2 = pd.json_normalize(credit['crew'].map(eval))
Df2
```

```python
Departament = []
Job = []
Name = []
for i in range(435):
    dep = 'department' + str(i)
    job = 'job' + str(i)
    nam = 'name' + str(i)
    Departament.append(dep)
    Job.append(job)
    Name.append(nam)
    Df2[dep] = pd.json_normalize(Df2[i])['department']
    Df2[job] = pd.json_normalize(Df2[i])['job']
    Df2[nam] = pd.json_normalize(Df2[i])['name']
```

```python
col = list(range(434))
Df2.drop(columns=col, inplace=True)
```

```python
Df2['departament'] = Df2[Departament].apply(lambda x: ',' .join(x.dropna().astype(str)),axis=1)
Df2['job'] = Df2[Job].apply(lambda x: '' .join(x.dropna().astype(str)),axis=1)
Df2['name_director'] = Df2[Name].apply(lambda x: ',' .join(x.dropna().astype(str)), axis=1)
```

```python
Df2.drop(columns=Departament, inplace=True)
Df2.drop(columns=Job, inplace=True)
Df2.drop(columns=Name, inplace=True)
```

```python
credit.drop(columns=['crew','cast'], inplace=True)
```

```python
Df_credit = pd.concat([credit,Df1,Df2], axis = 1)
```
